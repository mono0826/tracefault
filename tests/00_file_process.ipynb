## FileReader 数据类

```python
from backend.pipelines.file_reader import FileReader

data_path = os.getcwd() + "/../data/test/"
file_path = os.path.join(data_path, "test.pdf")

reader = FileReader()
Document = reader.read_files(file_paths=file_path)

print(Document)
```

## TextChunk 切分

```python
from backend.pipelines import ChineseTextChunker

spliter = ChineseTextChunker()

chunks = spliter.chunk_documents(Document)
print(f"len = {len(chunks)}")
print(f"chunk[0] = {chunks[3]}")
```

## DocumentProcess 类

```python
from backend.pipelines.document_processor import DocumentProcessor

processor = DocumentProcessor()
file_chunks = processor.process(file_paths=file_path)

print(f"文件数: {len(file_chunks)}")
print(f"返回值类型: {type(file_chunks)}")

total_chunks = 0
for file_name, chunks in file_chunks:
    total_chunks += len(chunks)
    print(f"  {file_name}: {len(chunks)} 个 Chunk")
    if chunks:
        print(f"    └ 首个 chunk: chunk_id={chunks[0].chunk_id[:12]}... content={chunks[0].content[:40]}...")

print(f"\n✓ 总文件: {len(file_chunks)}, 总 Chunk: {total_chunks}")
print("✓ process() 返回值格式正确")
```
